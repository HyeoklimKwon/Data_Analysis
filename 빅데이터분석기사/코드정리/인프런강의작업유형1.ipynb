작업유형1 문제2

데이터 첫번쨰 행부터 70%까지의 데이터를 추출한 후 

distance의 결측값을 distance의 중앙값으로 대체하고 

결측값 전처리 전과 후의 표준편차를 비교

전/후 결과값 print 출력



```python
import numpy as np
```

```python
df = sns.load_dataset('planets')
```

```python
df.head()
```

1.  70% 까지 데이터 추출

```python
len_df  = int(len(df) *0.7 )
print(len(df))
df = df[:len_df]
print(len(df))
```

2. distance 결측값 대체 하고 표준편차 비교

```python
df2 = df.copy() # 결측값 비교를 위해 얕은 복사 실행 
print(sum(df2['distance'].isnull())) # 결측치가 몇 개 있는 지 확인
df2['distance']= df2['distance'].fillna(df2['distance'].median()) # fillna를 이용하여 결측치를 중앙값으로 대체
print(sum(df2['distance'].isnull())) # 결측치가 몇 개 있는 지 확인
```

```python
print(np.std(df['distance']))
print(np.std(df2['distance']))
```

결측치 비교 결과, 결측치를 대체 했을 경우, 표준편차가 줄어들었다.

작업유형1 문제3

orbital_period열의 이상치를 IQR 기법으로 제거하여 이상치들의 합을 계산

IQR 기법 

IQR 방식은 사분위(Quantile) 개념으로부터 출발한다. 

전체 데이터들을 오름차순으로 정렬하고, 정확히 4등분(25%, 50%, 75%, 100%)으로 나눈다.

여기서 75% 지점의 값과 25% 지점의 값의 차이를 IQR이라고 한다.

이 IQR에 1.5를 곱해서 75% 지점의 값에 더하면 최댓값, 25% 지점의 값에서 빼면 최솟값으로 결정한다.

이 때, 결정된 최댓값보다 크거나 최솟값보다 작은 값을 이상치 라고 간주한다:.

```python
df = sns.load_dataset('planets')
df.head()
```

1. 이상치 범위 구하기

```python
Q25 = df['orbital_period'].quantile(0.25)
Q75 = df['orbital_period'].quantile(0.75)
print(Q25, Q75)
```

```python
IQR = Q75 - Q25
print(IQR)
min_value = Q25 - 1.5*IQR
max_value = Q75 + 1.5*IQR
print(min_value)
print(max_value)
```

2. 범위를 이용하여 이상치들의 합 계산하기

```python
outliers = df[(df['orbital_period'] <= min_value) | (df['orbital_period'] >= max_value) ]
outliers.head()
```

```python
print(len(df))
print(len(outliers))
```

```python
print(outliers['orbital_period'].sum())
```
